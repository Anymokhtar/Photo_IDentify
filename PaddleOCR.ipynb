```python
from paddleocr import PaddleOCR, draw_ocr
from PIL import Image

def inverse_arabic_text(text):
    arabic_chars = set("اأإآبتثجحخدذرزسشصضطظعغفقكلمنهويى")
    reversed_text = ""

    for char in text:
        if char in arabic_chars:
            reversed_text = char + reversed_text
        else:
            reversed_text += char

    return reversed_text

ocr = PaddleOCR(use_angle_cls=True, lang='ar', rec_image_inverse=False)
result = ocr.ocr(img_path, cls=True)

# Afficher les résultats inversés
for res in result:
    for line in res:
        box = line[0]
        text = line[1][0]
        confidence = line[1][1]

        inverted_text = inverse_arabic_text(text)

        print(f"Box: {box}, Texte inversé: {inverted_text}, Confiance: {confidence}")

# Dessiner le résultat
image = Image.open(img_path).convert('RGB')
boxes = [line[0] for line in result[0]]
txts = [inverse_arabic_text(line[1][0]) for line in result[0]]
scores = [line[1][1] for line in result[0]]
im_show = draw_ocr(image, boxes, txts, scores, font_path='/content/LateefRegOT.ttf')
im_show = Image.fromarray(im_show)
im_show.save('result2.jpg')
```

```python
from paddleocr import PaddleOCR, draw_ocr
from PIL import Image

def inverse_arabic_text(text):
    arabic_chars = set("اأإآبتثجحخدذرزسشصضطظعغفقكلمنهويى")
    reversed_text = ""

    for char in text:
        if char in arabic_chars:
            reversed_text = char + reversed_text
        else:
            reversed_text += char

    return reversed_text

def process_image(img_path):
    # Utiliser PaddleOCR pour détecter le texte
    ocr = PaddleOCR(use_angle_cls=True, lang='ar', rec_image_inverse=False)
    img_path = '/content/Carte_identité_électronique_marocaine_(2020,_recto).png'
    result = ocr.ocr(img_path, cls=True)

    # Afficher les résultats inversés
    for res in result:
        for line in res:
            box = line[0]
            text = line[1][0]
            confidence = line[1][1]

            inverted_text = inverse_arabic_text(text)

            print(f"Box: {box}, Texte inversé: {inverted_text}, Confiance: {confidence}")

    # Dessiner le résultat
    image = Image.open(img_path).convert('RGB')
    boxes = [line[0] for line in result[0]]
    txts = [inverse_arabic_text(line[1][0]) for line in result[0]]
    scores = [line[1][1] for line in result[0]]
    im_show = draw_ocr(image, boxes, txts, scores, font_path='/content/LateefRegOT.ttf')
    im_show = Image.fromarray(im_show)
    im_show.save('result.jpg')

# Chemin de l'image à traiter
img_path = 'chemin/vers/votre/image.jpg'

# Appeler la fonction pour traiter l'image
process_image(img_path)
```
